### UNION


```python
somf_data_union_defected = pd.read_csv('./Data/SIT/ai_somf_defect_sit_202006242.csv', na_values = '0')
somf_data_union_defected

somf_data_pre_wDefect = somf_data_pre.merge(somf_data_union_defected, left_on='ORDER_SEQ_ID', right_on='ORDER_SEQ_ID', how='left', sort=True)
somf_data_pre_wDefect.head()
```

### Indivisual defect add

```python
somf_data_received = pd.read_csv('./ai_data_somf_received_20200622.csv', na_values = '0')
somf_data_received.head()
somf_data_accepted = pd.read_csv('./ai_data_somf_accept_20200622.csv', na_values = '0')
somf_data_accepted.head()

concat_df = [somf_data_accepted, somf_data_received]
somf_labeled_defect = pd.concat(concat_df)
somf_labeled_defect

somf_data_pre_wDefect = somf_data_pre.merge(somf_labeled_defect, left_on='ORDER_SEQ_ID', right_on='ORDER_SEQ_ID', how='left', sort=True)
somf_data_pre_wDefect.head()
```

### Column 정리

```python
required_columns = ['ORDERNUMBER', 'ORRE', 'PRDI', 'LORD', 'PHRD', 'RECO',
       'WODI', 'REAC', 'JUJO', 'FUCO', 'SWBO', 'MAUP', 'IDMS', 'MEIN', 'CEMS',
       'MCID', 'DELS', 'VOIP', 'ORCN', 'FUCF', 'LABEL']
somf_data_final = somf_data_pre_wDefect[required_columns]
somf_data_final.drop_duplicates(inplace=True)
```

### PNA Data Loading

```python
pna_data = pd.read_csv('./ai_data_pa_20200622.csv')
pna_data.head()
```

```python
pna_data_pre = pna_data.fillna('0')
```

```python
pna_data_pre['MAX(POD.FIRST_CRET_DT)'] = pd.to_datetime(pna_data_pre['MAX(POD.FIRST_CRET_DT)'], format='%y/%m/%d')
pna_data_pre['MAX(POD.LAST_CHG_DT)'] = pd.to_datetime(pna_data_pre['MAX(POD.LAST_CHG_DT)'], format='%y/%m/%d')
pna_data_pre.drop(['MAX(POD.CUST_ORDR_VER_NO)'], axis =1,inplace=True) 
```

```python
pna_data_pre.head()
```

### 오늘의 쿼리
