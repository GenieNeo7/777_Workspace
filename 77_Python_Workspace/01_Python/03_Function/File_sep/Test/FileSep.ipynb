```python
newList = []

for i in test.values:
    tempList = i.tolist()
    if('ktFixedCommon'in i[14]):
        if('PUBLISH' in i[14].upper()):
                tempList.append('PNA_ORRE_Interaction')
        elif('DELAY' in i[14].upper()):
                tempList.append('PNA_PRDI_Interaction')    
        elif('CONTROLLER' in i[14].upper()):
                tempList.append('Step_Adjust')            
        elif('ktFixedCommon\resources\Functions\PnA\ProvisioningDelay' in i[14].upper()):
                tempList.append('PNA_PRDI_Interaction')    
        elif('ktFixedCommon\resources\Functions\PnA\ProvisioningDelay' in i[14].upper()):
                tempList.append('PNA_PRDI_Interaction')            
        elif('ConditionalCompensation' in i[14].upper()):
                tempList.append('Revision_Management')   
        elif('DATADICTIONARY' in i[14].upper()):
                tempList.append('Order_Data_Management')           
        else:
                tempList.append('Workflow_Execution')
    elif('ktFixedDesignAndAssignManagement'in i[14]):
        if('PHYSICAL' in i[14].upper()):
                tempList.append('PNA_PHRD_Interaction')
        elif('LOGICAL' in i[14].upper()):
                tempList.append('PNA_LORD_Interaction')      
        else:
                tempList.append('Workflow_Execution')    
    elif('ktFixedFieldWorkManagement'in i[14]):
        if('SWITCH' in i[14].upper()):
                tempList.append('PNA_SWBO_Interaction')
        elif('JUMPER' in i[14].upper()):
                tempList.append('PNA_JUJO_Interaction')      
        elif('WORKFORCE' in i[14].upper()):
                tempList.append('PNA_WODI_Interaction')      
        elif('FACILITY' in i[14].upper()):
                tempList.append('PNA_RECO_Interaction')                      
        else:
                tempList.append('Workflow_Execution')    
                
    elif('FulfillmentConfirmation'in i[14]):
        if('FULFILLMENT' in i[14].upper()):
                tempList.append('PNA_FUCO_Interaction')               
        else:
                tempList.append('Workflow_Execution')    

    elif('FulfillmentConfirmation'in i[14]):
        if('FULFILLMENT' in i[14].upper()):
                tempList.append('PNA_FUCO_Interaction')               
        else:
                tempList.append('Workflow_Execution')                    
                
    elif('ktFixedFulfillmentPatternMap 'in i[14]):
        if('DEPENDENCY' in i[14].upper()):
                tempList.append('Dependency_Management')
        elif('model\orchestrationModel\productSpec' in i[14]):
                tempList.append('ProductSpec_Management')      
        elif('ASSOCIATION' in i[14].upper()):
                tempList.append('ServiceCatalog_Management')     
        else:
                tempList.append('WorkflowRule_Management')     
    elif('DICTIONARY'in i[14].upper()):
        tempList.append('Order_Data_Management')
#     elif('ktFixedDesignAndAssignManagement'in i[14].upper()):
#         tempList.append('Order_Data_Management')        
    newList.append(tempList)
#     print(i[14])

df_test = pd.DataFrame(newList)
```
